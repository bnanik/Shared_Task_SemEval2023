```python
import pandas as pd
```

```python
def read_data(self,filename):
        sentences = []
        with open(filename, 'r', encoding='UTF-8') as f:
            sentence = []
            for line in f:
                line = line.strip()
                if len(line) != 0:
                    sentence.append((line.split(',')[4], line.split(',')[1]))   # getting text and label columns
                elif len(sentence) != 0:
                    sentences.append(sentence)
                    sentence = []
        self.tags = list(set(word_label[1] for sent in sentences for word_label in sent))
        return sentences 

```

```python
label_to_ids = {'not sexist':0,'sexist':1}
st_file="../Data/clean_data_v2.csv"
data=pd.read_csv(st_file)
data['label_sexist']=data['label_sexist'].map(label_to_ids)
data=data.filter(items=["text_clean_final","label_sexist"],axis=1)
data.columns=['text','label']

```
